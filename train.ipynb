```python
import emoji
import re
def clean_text(text):
    text = str(text)

    # 🔹 Convert emojis to text
    text = emoji.demojize(text)
    #text = text.replace("_", " ").replace(":", "")
    text = text.replace("_", " ")
    text = text.replace(":", " ")

    # 🔹 Lowercase
    text = text.lower()

    # 🔹 Remove URLs & mentions
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"@\w+", "", text)

    # 🔹 Remove extra spaces
    text = re.sub(r"\s+", " ", text)

    return text.strip()
```

#Load both datasets

```python
import pandas as pd
from IPython.display import display

train_df = pd.read_csv("train_CG.csv")
test_df  = pd.read_csv("dev_CG.csv")

train_df['clean_text'] = train_df['Text'].apply(clean_text)
test_df['clean_text']  = test_df['Text'].apply(clean_text)

display(train_df[['Text', 'clean_text']].head())
display(test_df[['Text', 'clean_text']].head())
```

```python
train_df
```

```python
#TF-IDF → converts text to numbers
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(token_pattern=r"(?u)\S+")  #lowercase=True,     #ngram_range=(1,2),  max_features=6000
X_train = tfidf.fit_transform(train_df['clean_text'])
y_train = train_df['Label']
X_test  = tfidf.transform(test_df['clean_text'])
y_test  = test_df['Label']
```

```python
features = tfidf.get_feature_names_out()
```

```python
features[1900:2000]
```

```python
#Train the model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
#Evaluate on test data
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

CONFUSION MATRIX


```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

plt.figure(figsize=(6, 5))
ConfusionMatrixDisplay.from_predictions(
    y_test,
    y_pred,
    cmap="Reds",
    xticks_rotation="vertical"
)
plt.title("Confusion Matrix")
plt.show()
```

```python
actual_counts = pd.Series(y_test).value_counts()
predicted_counts = pd.Series(y_pred).value_counts()

df = pd.DataFrame({
    "Actual": actual_counts,
    "Predicted": predicted_counts
}).fillna(0)

df.plot(
    kind="bar",
    figsize=(7, 5),
    color=["blue", "orange"]
)

plt.xlabel("Class Label")
plt.ylabel("Count")
plt.title("Actual vs Predicted Labels (Bar Graph)")
plt.xticks(rotation=45)
plt.show()
```

SVM

```python
from sklearn.svm import SVC

# Initialize and train the SVM model
# Using a linear kernel for simplicity and often good performance with TF-IDF
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import pandas as pd

# Make predictions on the test data
y_pred_svm = svm_model.predict(X_test)

# Evaluate the SVM model
print("SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
print("\nSVM Classification Report:\n", classification_report(y_test, y_pred_svm))

# Display Confusion Matrix for SVM
plt.figure(figsize=(6, 5))
ConfusionMatrixDisplay.from_predictions(
    y_test,
    y_pred_svm,
    cmap="Blues",
    xticks_rotation="vertical"
)
plt.title("SVM Confusion Matrix")
plt.show()
```

```python
# Display Actual vs. Predicted Labels (Bar Graph) for SVM
actual_counts_svm = pd.Series(y_test).value_counts()
predicted_counts_svm = pd.Series(y_pred_svm).value_counts()

df_svm_counts = pd.DataFrame({
    "Actual": actual_counts_svm,
    "Predicted": predicted_counts_svm
}).fillna(0)

df_svm_counts.plot(
    kind="bar",
    figsize=(7, 5),
    color=["green", "purple"]
)

plt.xlabel("Class Label")
plt.ylabel("Count")
plt.title("SVM Actual vs Predicted Labels (Bar Graph)")
plt.xticks(rotation=45)
plt.show()
```
